```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Data/players.csv")
```

```python
df.head(3)
```

```python
df["min"]=df["TotalGold"]/df["gold_per_m"]
df["kill/min"]=df["kill"]/df["min"]
df["assist/min"]=df["assist"]/df["min"]
```

```python
df['league'].unique()
list_param = ["csm", "gold_per_m", "goldshare", "min"]
role_stats = df.groupby(["league","role"]).mean()[list_param]
role_player_stats = df.groupby(["league","role","name_player"]).mean()[list_param]
role_player_percentage = 100*(role_player_stats -role_stats)/role_stats
role_stats
```

```python
role_stats
```

```python
role_player_percentage
roles = ['Top', 'Jungle', 'Mid', 'Bot', 'Support']
for league in df['league'].unique():
    fig, axes = plt.subplots(2, 3, figsize=(15,10))
    for i, role in enumerate(roles):
        ax = axes[int(i/3),i%3]
        role_player_percentage.loc[(league, role), : ].plot(ax=ax,kind='bar')
        #role_player_percentage.loc[((slice(None),role ), : ].plot(ax=ax,kind='bar')
        ax.title.set_text(league+" "+role)
        ax.set_ylabel('Distance to mean(%)')
        ax.set_xlabel('')
        ax.get_legend().remove()

    fig.legend(axes, labels=list_param,bbox_to_anchor=(0.78, 0.47),
               fancybox=True, shadow=True, prop={'size': 11})
    fig.tight_layout()
    fig.delaxes(axes[1,2])
    fig.savefig(f"Pictures/{league}")
```

```python
role_player_percentage.loc[("LCK", "Jungle"), : ].corr()
```

```python
fig, axes = plt.subplots(4, 5, figsize=(30,20))
for j, league in enumerate(df['league'].unique()):
    for i, role in enumerate(roles):
        ax = axes[j,i]
        #sns.heatmap(df.groupby(["league","role"])[["csm", "gold_per_m", "goldshare","win", "min"]].corr().loc[(league, roles), : ], cmap='seismic', square=True, annot=True, vmin=-1, vmax=1, ax=ax)
        sns.heatmap(role_player_percentage.loc[(league, role), : ].corr(), cmap='seismic', square=True, annot=True, vmin=-1, vmax=1, ax=ax)
        ax.title.set_text(league+" "+role)
fig.tight_layout()
fig.savefig(f"Pictures/heatmap")
```
